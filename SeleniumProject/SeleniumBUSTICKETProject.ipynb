```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

## Chrome Browser
# drive = webdriver.Chrome()

# drive.get('https://www.google.com.tw')


#driver_path = "C:\work\\anaconda\TestProject\chromedriver.exe"

## Brave Browser
brave_path = "C:\Program Files\BraveSoftware\Brave-Browser\Application/brave.exe"

service = Service(executable_path='./chromedriver.exe')
options = webdriver.ChromeOptions()
options.binary_location = brave_path
browser= webdriver.Chrome(service=service, options=options)

##購票網站
browser.get("https://order.kingbus.com.tw/ORD/ORD_M_1060_Anounce_OrderGB.aspx")

#定義相關參數
id = 'M122602563'
phone_num = '0912648699'
start_step = 'A19'
end_step = 'F20'
start_date = '2024/07/05'
start_time = '21'
end_date = '2024/07/07'
end_time = '13'

# 定位搜尋框
#element = browser.find_element('id','APjFqb')
# 傳入字串
#element.send_keys('Selenium Python')
browser.implicitly_wait(3)

#點選同意按鈕
button = browser.find_element('id','ctl00_ContentPlaceHolder1_btnOK')
button.click()

#輸入基本資料
element_id = browser.find_element('id','ctl00_ContentPlaceHolder1_txtCustomer_ID')
element_id.send_keys(id)
element_phone = browser.find_element('id','ctl00_ContentPlaceHolder1_txtPhone')
element_phone.send_keys(phone_num)

#輸入完資料後點選下一步
button_next = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_btnStep1_OK')
button_next.click()

##等待5秒


##選擇班次
##起點
element_start_select = Select(browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_ddlStation_ID_From'))
element_start_select.select_by_value(start_step)

##等待5秒
sleep(0.5)

#迄點
element_end_select = Select(browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_ddlStation_ID_To'))
element_end_select.select_by_value(end_step)


#起點日期
element_start_date = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_txtAOut_Dt')
element_start_date.send_keys(start_date)
#起點時間
element_start_hour = Select(browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_ddlAHour'))
element_start_hour.select_by_value(start_time)

#迄點日期
element_end_date = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_txtBOut_Dt')
element_end_date.send_keys(end_date)
#迄點時間
element_end_hour = Select(browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_ddlBHour'))
element_end_hour.select_by_value(end_time)

#下一步
element_next_button  = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_btnStep2_OK')
element_next_button.click()



#選班次
element_item_button =  browser.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_grdAList"]/tbody/tr[2]/td[2]/input')
element_item_button.click()

sleep(1)
element_item_button2 = browser.find_element(By.XPATH,'//*[@id="ctl00_ContentPlaceHolder1_grdBList"]/tbody/tr[2]/td[2]/input')
element_item_button2.click()
sleep(0.5)

#確定
element_enter_button = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_btnStep3_OK')
element_enter_button.click()

#電腦配位
element_auto_button1 = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_ckbAComputer')
element_auto_button1.click()

element_auto_button2 = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_ckbBComputer')
element_auto_button2.click()

##購買票數

element_ticket_num_1 = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_rdoATot_Count_0')
element_ticket_num_1.click()

element_ticket_num_2 = browser.find_element(By.ID,'ctl00_ContentPlaceHolder1_rdoBTot_Count_0')
element_ticket_num_2.click()

ticks_num = 1



#browser.close()

```
